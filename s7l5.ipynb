```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('C:/Users/leand/Desktop/dataset_climatico.csv')
table_clean = df.dropna().drop_duplicates()


df['temperatura_media_z'] = stats.zscore(df['temperatura_media'])
df['precipitazioni_z'] = stats.zscore(df['precipitazioni'])
df['umidita_z'] = stats.zscore(df['umidita'])
df['velocita_vento_z'] = stats.zscore(df['velocita_vento'])

print(df[['temperatura_media_z', 'precipitazioni_z', 'umidita_z', 'velocita_vento_z']])
desc_stats = df[['temperatura_media_z', 'precipitazioni_z', 'umidita_z', 'velocita_vento_z']].describe()
print(desc_stats)
plt.figure(figsize=(12, 6))
for col in desc_stats.columns:
    sns.histplot(df[col], kde=True, label=col)
plt.xlabel('Valore normalizzato Z-score')
plt.ylabel('Frequenza')
plt.title('Distribuzione delle variabili normalizzate')
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['temperatura_media_z', 'precipitazioni_z', 'umidita_z', 'velocita_vento_z']])
plt.xlabel('Variabile')
plt.ylabel('Valore normalizzato Z-score')
plt.title('Box Plot delle variabili normalizzate')
plt.show()
```
