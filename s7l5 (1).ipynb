```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('C:/Users/leand/Desktop/dataset_climatico.csv')
table_clean = df.dropna().drop_duplicates()


df['temperatura_media_z'] = stats.zscore(df['temperatura_media'])
df['precipitazioni_z'] = stats.zscore(df['precipitazioni'])
df['umidita_z'] = stats.zscore(df['umidita'])
df['velocita_vento_z'] = stats.zscore(df['velocita_vento'])

print(df[['temperatura_media_z', 'precipitazioni_z', 'umidita_z', 'velocita_vento_z']])
desc_stats = df[['temperatura_media_z', 'precipitazioni_z', 'umidita_z', 'velocita_vento_z']].describe()
print(desc_stats)
plt.figure(figsize=(12, 12))  # Aumenta la dimensione della figura per ospitare tutti i grafici

# Grafico 1 (prima riga, prima colonna)
plt.subplot(2, 2, 1)
sns.histplot(df['temperatura_media_z'], label='temperatura_media_z', kde=True)
plt.xlabel('Valore normalizzato Z-score')
plt.ylabel('Frequenza')
plt.title('Distribuzione temperatura_media_z')

# Grafico 2 (prima riga, seconda colonna)
plt.subplot(2, 2, 2)
sns.histplot(df['precipitazioni_z'], label='precipitazioni_z', kde=True)
plt.xlabel('Valore normalizzato Z-score')
plt.ylabel('Frequenza')
plt.title('Distribuzione precipitazioni_z')

# Grafico 3 (seconda riga, prima colonna)
plt.subplot(2, 2, 3)
sns.histplot(df['umidita_z'], label='umidita_z', kde=True)
plt.xlabel('Valore normalizzato Z-score')
plt.ylabel('Frequenza')
plt.title('Distribuzione umidita_z')

# Grafico 4 (seconda riga, seconda colonna)
plt.subplot(2, 2, 4)
sns.histplot(df['velocita_vento_z'], label='velocita_vento_z', kde=True)
plt.xlabel('Valore normalizzato Z-score')
plt.ylabel('Frequenza')
plt.title('Distribuzione velocita_vento_z')

plt.tight_layout()  # Aggiunge spazi tra i grafici per migliorare la leggibilit
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['temperatura_media_z', 'precipitazioni_z', 'umidita_z', 'velocita_vento_z']])
plt.xlabel('Variabile')
plt.ylabel('Valore normalizzato Z-score')
plt.title('Box Plot delle variabili normalizzate')
plt.show()
```
